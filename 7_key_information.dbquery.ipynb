```sql
CREATE OR REPLACE TABLE IDENTIFIER(:catalog || '.' || :schema || '.key_info') AS
  SELECT 
    f.path,
    CONCAT_WS('\n', 
      'VENDOR_NAME', f.vendor_name,
      '\n FILE_NAME', f.file_name,
      '\n OTHER_VENDOR_FILES', array_join(f.other_vendor_files, ', '),
      '\n PREAMBLE', SUBSTRING(f.preamble,0,500),
      '\n METADATA', CONCAT( 
        '\n Agreement Name: ', m.agreement_name,
        '\n Agreement Title: ', m.agreement_title,
        '\n Effective Date: ', m.effective_date,
        '\n Expiry Date: ', m.expiry_date,
        '\n Project Number: ', m.project_number,
        '\n Is Evergreen Agreement: ', m.is_evergreen_agreement,
        '\n Contract Category: ', m.contract_category
      ),
      '\n REFERENCES', r.combined_references
    ) AS key_info
  FROM shm.contracts.flat f
  LEFT JOIN (SELECT path, combined_references FROM shm.contracts.references) r
  ON f.path = r.path
  LEFT JOIN shm.contracts.metadata m
  ON f.path = m.path
```
