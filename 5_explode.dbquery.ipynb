```sql
SELECT 
  *,
  from_json(
    get_json_object(classification, '$.MasterAgreements'),
    'array<struct<FileName:string,RelatedDocuments:struct<Amendments:array<struct<FileName:string,Date:string,Sequence:string>>,RateSheets:array<string>,OtherRelatedFiles:array<string>>>>'
  ) AS agreements,
  from_json(
    get_json_object(classification, '$.UnclassifiedFiles'),
    'array<struct<FileName:string,Title:string>>'
  ) AS unclassified_files
FROM IDENTIFIER(:catalog || '.' || :schema || '.classified')
```
